```python
# Load the cleaned orders data from the previous step
import pandas as pd
import matplotlib.pyplot as plt
# Specify the full path to your file
df = pd.read_csv('C:/Users/alfateh/Documents/python practice/cleaned_orders_final.csv')
df.head()
```

```python
#convert order_date to datetime
df['ShippedDate'] = pd.to_datetime(df['ShippedDate'])
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
# set the figure size for better visibility/Readability
plt.figure(figsize=(10, 6))

#Group data by country and calculate the average shipping cost per country
#Then sort the results for a cleaner horizontal bar chart
df.groupby('ShipCountry')['ShippingCost'].mean().sort_values().plot(kind='barh', color='skyblue')

#set the title of the plot 
plt.title('Average Shipping Cost by Country')
#label the axes
plt.xlabel('Average Shipping Cost')
plt.ylabel('Country')
#show  gridline for better readability
plt.grid(True, axis='x')
#automatically adjust the layout to prevent overlap/cutoff
plt.tight_layout()
#display the plot
plt.show()

```

```python
#Visualizatio 2 : Delivery Days By Destination 
#set the figure size for better visibility/Readability
plt.figure(figsize=(10, 6))
#plot a Histogram of delivery days 
#drop any missing values to avoid errors in plotting
df['DeliveryDays'].dropna().plot(kind='hist', bins=20, color='#649532ED', edgecolor='black')
#set the title of the plot
plt.title('Distribution of Delivery Days')
#label the axes
plt.xlabel('Delivery Days')
plt.ylabel('Frequency')
#show gridlines for better readability
plt.grid(True)
#automatically adjust the layout to prevent overlap/cutoff
plt.tight_layout()
#Render the plot
plt.show()
```

```python
#Visualization 3: Delivery Status Count
#set the figure size for better visibility/Readability
plt.figure(figsize=(6, 6))
#plot a pie chart of delivery status counts
df['DeliveryStatus'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=160, colors=['#79AE6F','#9FCB98', '#F2EDC2'])
#set the title of the plot
plt.title('Delivery Status Distribution')
#equal aspect ratio ensures that pie chart is circular
plt.axis('equal')
#automatically adjust the layout to prevent overlap/cutoff
plt.tight_layout()
#Render the plot    
plt.show()
```

```python
#VISUALIZATION 4: Monthly Order Volume
#Extract the month from the order date format: YYYY-MM
df["Months"] = df["OrderDate"].dt.to_period("M")

#Group data by month and count the number of orders per month
monthly_orders = df.groupby("Months").size()
#print(monthly_orders)

#set the figure size for better visibility/Readability
plt.figure(figsize=(12, 6))

#plot a line chart of monthly order volume
monthly_orders.plot(kind='line', marker='o', color='purple')
#set the title of the plot
plt.title('Monthly Order Volume')
#label the axes
plt.xlabel('Month')
plt.ylabel('Number of Orders')
#show gridlines for better readability
plt.grid(True)
#automatically adjust the layout to prevent overlap/cutoff
plt.tight_layout()
#Render the plot
plt.show()

```
